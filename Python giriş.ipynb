```python
def isim_gizleme(isim_soyad):
  ad_soyad_list = isim_soyad.split()
  isim_gizleme_str=""
  for isim in ad_soyad_list:
   isim_gizleme_str=isim[:2]+"****"
  return isim_gizleme_str
```
