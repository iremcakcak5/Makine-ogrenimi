```python
# Ülkelerin ve yılların listesini alma
countries = list(df.columns[1:])
years = list(df.index)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(88,70))
bar_list = []
```

```python
#Her ülke için bir çubuk oluşturma
for i, country in enumerate(countries):
    bar = ax.barh(i, df[country][0], color=[plt.get_cmap('tab20')(j) for j in np.linspace(0, 1, len(countries))][i])
    bar_list.append(bar)
```

```python
#x-Y eksenleri sınır ayarlama
ax.set_xlim(0, df[countries[0]].max()*1.1)
ax.set_ylim(-1, len(countries))
```

```python
#Eksenleri ve grafiğe isimlendirme
ax.set_xlabel('Population')
ax.set_ylabel('Country')
ax.set_title('World Population Competition')
```

```python
import numpy as np
```

```python
ax.set_yticks(np.arange(len(countries)))
ax.set_yticklabels(countries, rotation='horizontal', fontsize=22) # Dikey olarak yazdırma, daha küçük yazı boyutu
plt.tight_layout(pad=3.0)
```

```python
#animasyon fonksiyonu
def animate(i):
    for j, bar in enumerate(bar_list):
        for rect in bar:
            rect.set_width(df[countries[j]][i])
```

```python
import matplotlib.animation as animation
```

```python
# animasyon yaratma
ani = animation.FuncAnimation(fig, animate, frames=len(years), interval=1000)
```

```python
#animasyonu mp4 dosyası olrak kaydetme
ani.save('world_population_competition.mp4')
```
